```python
# Pie Chart

import matplotlib.pyplot as plt

slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow= True,explode=(0,0.1,0,0),autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show()

# explode works by index i.e (0,0.1,0,0) means sleeping will not expand outwards, eating will expand by 0.1 and working and 
# playing won't expand outwards as well. Pie chart will start anti-clockwise. Autopct describes how much value to be taken after
# decimal point in percentage. %1.1f%% means only one value will be taken after decimal while calculating percentage.
```

```python
# Area Plot/ Stack Plot: multiple data series are stacked on top of each other to show their contribution to a total over time.

import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,10]

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'], labels = ['sleeping','eating','working','working'])
plt.xlabel('No. of Days')
plt.ylabel('Hours')
plt.title('Stack Plot')
plt.legend()
plt.show()
```
