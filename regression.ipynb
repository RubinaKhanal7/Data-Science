```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary Data.csv')
dataset=dataset.dropna()
# dataset.to_csv('new_dataset.csv')
dataset
```

```python
dataset = dataset[['Years of Experience','Salary']]
dataset
```

```python
X = dataset.iloc[:,0:1].values
X
```

```python
# .values converts into 2 dimensional numpy array
# iloc is index localizer. Here first pair represents rows and second pair represents column

Y = dataset.iloc[:,1:].values
```

```python
# X is independent variable and Y is dependent variable. test_size takes 20% data for testing and 80% for training. random_state
#  controls how the dataset is shuffled before splitting.
# If you set a fixed number (e.g., random_state=0), the shuffling will always be the same, so you get the same train-test split
# every time you run the code.
# If you don't set it (or use random_state=None), the shuffling will be random every time, so the train-test split will be 
# different each time you run the code.

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
# print(train_test_split(X,Y,test_size = 1/3,random_state = 0))
```

```python
print(len(Y_train),len(X_train))
print(len(Y_test),len(X_test))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()   # initialize linear regression model
model.fit(X_train,Y_train)   # this code is used for model training
```
